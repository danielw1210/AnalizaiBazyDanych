```python
from IPython.core.display import display
import numpy as np
import pandas as pd


# Wczytanie pliku z danymi billboard.csv z folderu z nieobrobionymi danymi ('Original Data')
data = pd.read_csv('../Original Data/billboard.csv')

# Wyświetlenie oryginalnych danych
display(data)
```

```python
# Utworzenie tabeli z danymi piosenek (rok produkcji, artysta, nazwa piosenki, gatunek muzyczny oraz czas trwania)
data1 = data[['artist.inverted','track','genre','time']]
data1.columns=["Artist",'Track','Genre','Time']

# Dodanie id w celu integracji ze sobą 2 tabel (z danymi piosenek oraz danymi rankingowymi)
idxs = [index for index, row in data1.iterrows()]
data1.insert(0,'song id',idxs)

# Wyświetlenie utworzonej tabeli
display(data1)

# Zapisanie tabeli do pliku csv do folderu Analysis Data, gdzie znajdują się przetworzone dane
data1.to_csv('../Analysis Data/SongsData.csv')
```

```python
# Utworzenie tabeli z danymi rankigowymi zmiennymi w czasie odnoszącymi się do piosenek o id zgodnymi z poprzednią tabelą
data2 = data.iloc[:,7:]

# Utworzenie pustego DataFrame
ranking_data = pd.DataFrame(columns=['song id','date','rank'])

# Dodawanie do DataFrame ranking_data rekordów z id piosenki, datą oraz pozycją w rankingu
for index, row in data2.iterrows():
    date = pd.to_datetime(data.iloc[index]['date.entered'])
    for column in row:
        if str(column) != 'nan':
            week_data = pd.DataFrame([[index, date, int(column)]], columns=['song id','date','rank'])
            ranking_data = pd.concat([ranking_data, week_data], ignore_index=True)
            date = date + pd.DateOffset(7)

            


# Wyświetlenie utworzonej tabeli
display(ranking_data)

# Zapisanie tabeli do pliku csv do folderu Analysis Data, gdzie znajdują się przetworzone dane
ranking_data.to_csv('../Analysis Data/RankingData.csv')

```
