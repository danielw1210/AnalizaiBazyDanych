### __Zadanie 3__ 

Zadanie miało na celu przypomnienie wiedzy na temat przedstawiania danych na wykresach z wykorzystaniem biblioteki matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt

# Pythonowa funkcja opisująca funkcję kwadratową
def function1(x):
    return x**2 + 5

# Utworzenie przedziałów argumentów zgodnie z instrukcją, każdy po 100 punktów
data1 = np.linspace(-1,1,100)
data2 = np.linspace(-6,6,100)
data3 = np.linspace(-0,5,100)

# Utworzenie wykresu dla pierwszego przedziału argumentów
plt.plot(data1, function1(data1))
plt.xlabel('Argumenty')
plt.ylabel('Wartośći')
plt.title('f(x) = x^2 + 5')
plt.suptitle('Argumenty z przedziału (-1,1)')
plt.tight_layout()
plt.grid()
plt.show()

# Utworzenie wykresu dla drugiego przedziału argumentów
plt.plot(data2, function1(data2))
plt.xlabel('Argumenty')
plt.ylabel('Wartośći')
plt.title('f(x) = x^2 + 5')
plt.suptitle('Argumenty z przedziału (-6,6)')
plt.tight_layout()
plt.grid()
plt.show()

# Utworzenie wykresu dla trzeciego przedziału argumentów
plt.plot(data3, function1(data3))
plt.xlabel('Argumenty')
plt.ylabel('Wartośći')
plt.title('f(x) = x^2 + 5')
plt.suptitle('Argumenty z przedziału (0,6)')
plt.tight_layout()
plt.grid()
plt.show()
```
